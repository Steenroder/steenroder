The coboundary matrix with an order of the basis that is the opposite of that of the order of the simplexwise filtration. I.e., from largest to smallest dimensions.

```python
def get_coboundary(simplexwise_filtration):
    '''
    Input: list of tuples - Output: np.array
    
    using mod 2 coefficients
    
    The order of rows and columns is last to first w/r to filtration order
    '''
    n = len(simplexwise_filtration)
    spx_filtration_idx = {tuple(v): idx for idx, v in enumerate(simplexwise_filtration)}
    bdry = np.zeros((n, n), dtype=np.bool)
    for idx, spx in enumerate(simplexwise_filtration):
        faces_idxs = []
        try:
            faces_idxs = [spx_filtration_idx[spx[:j]+spx[j+1:]] for j in range(len(spx))]
        except KeyError:
            pass
        bdry[faces_idxs,idx] = True

    return np.flip(bdry, axis=[0,1]).transpose()
```

```python
def pivot(column):
    
    try:
        return max(column.nonzero()[0])
    except ValueError:
        return None

def get_reduced_triangular(matrix):
    '''R = MT'''
    n = matrix.shape[1]
    reduced = np.array(matrix)
    triangular = np.eye(n, dtype=np.bool)
    for j in range(n):
        i = j
        while i > 0:
            i -= 1
            if not np.any(reduced[:,j]):
                break
            else:
                piv_j = pivot(reduced[:,j])
                piv_i = pivot(reduced[:,i])
                
                if piv_i == piv_j:
                    reduced[:,j] = np.logical_xor(reduced[:,i], reduced[:,j])
                    triangular[:,j] = np.logical_xor(triangular[:,i], triangular[:,j])
                    i = j
                    
    return reduced, triangular
```

```python
def get_barcode(reduced):
    '''barcoded ordered by second coordinate'''
    n = reduced.shape[0]
    pairs = []
    all_indices = []
    for j in range(n):
        if np.any(reduced[:,j]):
            i = pivot(reduced[:,j])
            pairs.append((i,j))
            all_indices += [i,j]
    
    for i in [i for i in range(n) if i not in all_indices]:    
        if not np.any(reduced[:,i]):
            pairs.append((i,np.inf))
    
    return pairs
```

```python
from itertools import combinations

# set -> array
def cochain_to_vector(cochain, simplex_to_index):
    n = len(simplex_to_index)
    nonzero_indices = [simplex_to_index[spx] for spx in cochain]
    vector = np.zeros(shape=(n,1), dtype=np.bool)
    vector[nonzero_indices] = True
    return vector

# array -> set
def vector_to_cochain(vector, reversed_filtration):
    return {reversed_filtration[i] for i in vector.nonzero()[0]}

# (int, array) --> array
def STSQ(k, vector):
    cocycle = vector_to_cochain(vector, reversed_filtration)
    answer = set()
    for pair in combinations(cocycle, 2):
        a, b = set(pair[0]), set(pair[1])
        if ( len(a.union(b)) == len(a)+k and 
        tuple(sorted(a.union(b))) in reversed_filtration ):
            a_bar, b_bar = a.difference(b), b.difference(a)
            index = dict()
            for v in a_bar.union(b_bar):
                pos = sorted(a.union(b)).index(v)
                pos_bar = sorted(a_bar.union(b_bar)).index(v)
                index[v] = (pos + pos_bar)%2
            index_a = {index[v] for v in a_bar}
            index_b = {index[w] for w in b_bar}
            if (index_a == {0} and index_b == {1} 
            or index_a == {1} and index_b == {0}):
                u = sorted(a.union(b))
                answer ^= {tuple(u)}

    return cochain_to_vector(answer, simplex_to_index)
```

```python
def get_pivots(matrix):
    n = matrix.shape[1]
    pivots = []
    for i in range(n):
        pivots.append(pivot(matrix[:,i]))
        
    return pivots


def reduce_vector(reduced, vector):
    '''
    input:  (n,m) array 
            (n, ) array
    
    output: (n, ) array
    '''
    num_col = reduced.shape[1]
    i = -1
    while i >= -num_col:
        if not np.any(vector):
            break
        else:
            piv_v = pivot(vector)
            piv_i = pivot(reduced[:,i])

            if piv_i == piv_v:
                vector[:,0] = np.logical_xor(reduced[:,i], vector[:,0])
                i = -1
            
            i -= 1
    
    return vector

def reduce_matrix(reduced, matrix):
    num_vector = matrix.shape[1]

    for i in range(num_vector):
        reduced_vector = reduce_vector(reduced, matrix[:, i:i+1])
        reduced = np.concatenate([reduced, reduced_vector], axis=1)
    return reduced[:, -num_vector:]
```
